```python
from ffnn import FFNN
import numpy as np
import pandas as pd
```

```python
ffnn = FFNN(layer_sizes=[5, 3, 1], activations=['tanh', 'linear'], weight_init='random_normal',init_params={'mean' : 0 , 'variance' : 0.05}, seed=42)
```

```python
data=pd.read_csv('fictive_dataset.csv')
```

```python
X=data.drop('label', axis=1)
```

```python
y=ffnn.forward_propagation(X)
```
